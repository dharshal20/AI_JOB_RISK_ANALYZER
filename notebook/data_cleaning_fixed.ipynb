```python
import pandas as pd

df = pd.read_csv('ds_salaries.csv')
df.head()
```

```python
df.head()
df.info()
df.describe()
df.columns
```

```python

df.drop_duplicates(inplace=True)


df.isnull().sum()
```

```python
df.fillna('Unknown', inplace=True)
```

```python
def ai_risk(role):
    role = str(role).lower()
    
    if 'analyst' in role:
        return 'Medium'
    elif 'engineer' in role or 'scientist' in role:
        return 'Low'
    else:
        return 'High'

df['ai_risk'] = df['job_title'].apply(ai_risk)
```

```python
def career_advice(role):
    role = str(role).lower()
    
    if 'scientist' in role or 'ml' in role:
        return 'High Growth Role'
    elif 'analyst' in role:
        return 'Upskill to ML/AI'
    else:
        return 'At Risk – Upskill Needed'

df['career_advice'] = df['job_title'].apply(career_advice)
```

```python
df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
```
